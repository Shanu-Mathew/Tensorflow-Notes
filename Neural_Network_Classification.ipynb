To visualize our Model's predictions, let's create a function plot_decision_boundary(),this function will:
1. Take in a trained model, features(X) and labels(y)
2. Create a meshgrid of the different X values
3. Make predictions across the meshgrid
4. Plot the predictions as well as a line between zones (where each unique class falls)


```python
def plot_decision_boundary(model,X,y):
  #Define the axis boundaries of the plot and create a meshgrid
  x_min,x_max=X[:,0].min()-0.1,X[:,0].max()+0.1
  y_min,y_max=X[:,1].min()-0.1,X[:,1].max()+0.1

  xx,yy=np.meshgrid(np.linspace(x_min,x_max,100),
                    np.linspace(y_min,y_max,100))

  #Create X value
  x_in =np.c_[xx.ravel(),yy.ravel()]

  #make predictions
  y_pred=model.predict(x_in)
  print(len(y_pred[0]))
  #Check for multi-class
  if(len(y_pred[0])>1):
    print("Doing multiclass Classification")
    # We have to reshape out prediction to get them ready for plotting
    y_pred=np.argmax(y_pred,axis=1).reshape(xx.shape)
  else:
    print("Doing binary classification")
    y_pred=np.round(y_pred).reshape(xx.shape)

  # Plot the decision boundary
  plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(),xx.max())
  plt.xlim(yy.min(),yy.max())
```

```python
#Check out the prediction
plot_decision_boundary(model=model_3,X=X,y=y)
```

### Learning Rate Callback
This is an extra piece of functionality to determine a better learning rate.

```python
#Set Random Seed
tf.random.set_seed(42)

#Creating a model
model_8=tf.keras.Sequential([
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

# Compile the model
model_8.compile(loss="binary_crossentropy",
                optimizer="Adam",
                metrics=["accuracy"])

# Create a learning rate callbaack
lr_scheduler=tf.keras.callbacks.LearningRateScheduler(lambda epoch:1e-4 * 10**(epoch/20))

#Fit the model (passing lr_scheduler callback)
history_8=model_8.fit(X_train,y_train, epochs=100,callbacks=lr_scheduler)
```

```python
pd.DataFrame(history_8.history).plot(figsize=(10,7),xlabel="epochs");
```

```python
lrs=1e-4 * (10 ** (tf.range(100)/20))
plt.figure(figsize=(10,7))
plt.semilogx(lrs,history_8.history["loss"])
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning Rate vs Loss")
```

The ideal learning rate is somewhere between the lowest point in the curve and the region where there is a sharp decrease in loss
